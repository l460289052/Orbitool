```python
import httpx

import pyteomics.mass
```

```python
mz = 294.08305 - 0.00054857990943
mz
```

```python
url = "https://www.chemcalc.org/chemcalc/em"
params = {'mfRange': 'C0-30 H0-40 N0-3 O0-20 [18O]0-1 [13C]0-2',
          'numberOfResultsOnly': False,
          'typedResult': False,
          'useUnsaturation': False,
          'minUnsaturation': 0,
          'maxUnsaturation': 50,
          'jcampBaseURL': 'http://www.chemcalc.org/service/jcamp/',
          'monoisotopicMass': mz,
          'jcampLink': False,
          # The 'jcamplink' returns a link to a file containing isotopic
          # distribution of retrieved molecular structure.
          'integerUnsaturation': False,
          'maxNumberRows': 100,
          # Ions/Radicals can have non-integer unsaturation
          'referenceVersion': '2013',
          'massRange': 0.5
          #              'minMass': -0.5,
          #              'maxMass': 0.5,
          }
```

```python
help(httpx.get)
```

```python
ret = httpx.get(url, params=params)
```

```python
ret.json()
```

```python
for r in ret.json()['results']:
    print(r['mf'], r['em'], r['ppm'])
```
